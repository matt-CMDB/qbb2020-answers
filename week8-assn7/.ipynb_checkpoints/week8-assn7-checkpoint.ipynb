# Clustering

Use SciPy

- Limit the dataset to genes with a median expression across samples of greater than zero.

- Apply a log2(FPKM + 0.1) transformation to the data.

- Cluster the data matrix for both genes and samples on their patterns of expression (so both the rows and columns of the matrix), and plot a heatmap of the gene expression data.

- Next create a dendrogram relating the samples to one another.

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf
from scipy.cluster.hierarchy import dendrogram, linkage, leaves_list
```

```python
df = pd.read_csv('all_annotated.csv', delimiter=',')

#get median values:
df['Median'] = df.loc[2:].median(axis=1)
df
```

```python
#remove genes with 0 FPKM values
FPKM = df[(df['Median'] > 0 ) == True]
#log2 transform:
logFPKM = FPKM.iloc[:,:2].join(np.log2(FPKM.iloc[:,2:] + 0.1)).reset_index().iloc[:,1:]

logFPKM
```

```python
#Cluster the data matrix for both genes and samples on their patterns of expression 
#(both the rows and columns of the matrix)

# first, get only the numeric data for clustering and make arrays
samples = np.array(list(logFPKM.columns)[2:-1])
names = np.array(logFPKM.iloc[:,:2])
exp_values = np.array(logFPKM.iloc[:,2:-1])
names
```

```python
# gene clustering - values and associated gene names
gene_cluster = exp_values[leaves_list(linkage(exp_values))]
name_cluster = names[leaves_list(linkage(exp_values))]

name_cluster
```

```python
# sample clustering: same idea, but transpose the table and use the column names

sample_cluster = gene_cluster.T[leaves_list(linkage(gene_cluster.T))].T
sname_cluster = samples[leaves_list(linkage(gene_cluster.T))]
sname_cluster
```

```python
# plot a heatmap of the expression data by gene with seaborn
import seaborn
plt.figure(figsize=(12,12)) # the default size is too small to see much
fig = seaborn.heatmap(gene_cluster)
fig.set_xticklabels(sname_cluster, rotation =45, ha='right') #clustering on two axes req use of sample name
fig.set_xlabel("Samples")
fig.set_ylabel("Genes")
plt.show()
```

```python
# dendrogram!
fig = plt.figure(figsize=(10, 6))
dgram = dendrogram((linkage(gene_cluster.T)), labels=samples, leaf_rotation = 45)
plt.show()
```

# Differential expression


```python
#First, we have to fix up the numbers for developmental stage and sex of the samples:

#transpose the df to make it easier to add "sex" and "dev_stage" columns

df = logFPKM.T

#start lists for sex and developmental stage
sex = [] 
dev_stage = []

# check through the index list 
# yes I'm sure there's a better way to do this, but...

for i in df.index:
    if 'female' in i:
        sex.append('F')
    elif 'male' in i:
        sex.append('M')
    else:
        sex.append("Sex")

for i in df.index:
    if '10' in i:
        dev_stage.append(float(10))
    elif '11' in i:
        dev_stage.append(float(11))
    elif '12' in i:
        dev_stage.append(float(12))
    elif '13' in i:
        dev_stage.append(float(13))
    elif '14' in i:
        dev_stage.append(float(14))
    else:
        dev_stage.append("dev_stage")

df['sex'] = sex
df['dev_stage'] = dev_stage

#now remove some extra rows
values = df[1:-1]

#name columns according to their genes
values.columns = values.iloc[0]
values = values.iloc[1:]
values
```

### Regression by developmental stage
Use ordinary least squares regression to test for genes that are differentially expressed across stages. Use the stage number as a numeric independent variable (10, 11, 12, 13, 14), and ignore the letter suffixes on day 14 (i.e., treat 14A, 14B, 14C, and 14D as equivalent).

```python
# I used this mysterious line of code to clean up my dataframes from duplicate columns 
df = values.loc[:,~values.columns.duplicated()]

#unfortunately there were still wierd things going on so I did this to make them into floats
transcripts = df.columns[0:-2]
df = df.iloc[:,:-2].astype(float)
#replace the missing non-float columns:
df['dev_stage'] = values['dev_stage']
df['sex'] = values['Sex']

df

#I probably lost some data but after too many hours of trying to figure out what's going on imma let it be
```

```python
#test case
temp_data = {'fpkm':df['CG14641'], 'stage':df['dev_stage'], 'sex':df['sex']}
temp_df = pd.DataFrame(temp_data)
model = smf.ols(formula = "fpkm ~ stage", data = temp_df)
sex_model = smf.ols(formula = "fpkm ~ stage + sex", data = temp_df)

results = model.fit() 
sex_results = sex_model.fit()

sex_results
```

```python
transcripts = df.columns[0:-2]


    
pvals = [] 
pvals_sex = [] 
bvals =[]

for transcript in transcripts:
    #make a temporary dataframe to feed values into ols model
    temp_data = {'fpkm':df[transcript], 'stage':df['dev_stage'], 'sex':df['sex']}
    temp_df = pd.DataFrame(temp_data)
    model = smf.ols(formula = "fpkm ~ stage", data = temp_df)
    sex_model = smf.ols(formula = "fpkm ~ stage + sex", data = temp_df)

    results = model.fit() 
    sex_results = sex_model.fit()

    pvals.append(results.pvalues[1])
    pvals_sex.append(sex_results.pvalues[2])
    bvals.append(sex_results.params[2])
```

```python
#add parameters to df
#first gotta transform it back
df = df.iloc[:,:-2]
df = df.T

df['stage_p'] = pvals
df['stage-sex_p'] = pvals_sex
df['bvals'] = bvals
df
```

###  QQ plot from p-values
2. Generate a QQ plot from the p-values.

3. Report the list of genes that exhibit differential expression by stage at a 10% false discovery rate.

4. Repeat the analysis while controlling for sex.

5. Report the list of genes that exhibit differential expression by stage at a 10% false discovery rate while controlling for sex.

6. Compare the lists–what is the percentage overlap with and without sex as a covariate?

7. Generate a volcano plot of the differential expression, with sex as a covariate, results ( -log10(p-value) on the y-axis, beta on the x axis). Color the significant points in a different color.

```python
#copied from the old QQ plots assignment
df_sorted = df.sort_values(by = "stage_p")
df_sorted['logP'] = -1 * np.log10(df_sorted['stage_p'])
df_sorted['uniform_points'] = range(0, len(df_sorted))
df_sorted['uniform_pval'] = (df_sorted['uniform_points'] + 1) / len(df_sorted)
df_sorted['uniform_logP'] = -1 * np.log10(df_sorted['uniform_pval'])

df_sorted
```

```python
fig, axs = plt.subplots(1,2)
axs[0].scatter(df_sorted["uniform_logP"], df_sorted["logP"])
axs[0].plot([8,0], [8,0], color = "black")
axs[0].set_xlim([0, 8])
axs[0].set_ylim([0, 10])
axs[0].set_xlabel("Expected -log10(p-value)")
axs[0].set_ylabel("Observed -log10(p-value)")
axs[0].set_title("QQ plot: dev stage")

plt.show()
# looks like the vast majority of genes are subject to differential regulation based on dev stage
```

```python
#BUT... now use Bonferoni correction for false discovery rate

#using a 10% false discovery rate
fdr = 0.1/5556
df_fdr = df[df['stage_p']<fdr]
df_fdr
```

```python
#now on to sex...


df_sorted_psex = df.sort_values(by = "stage-sex_p")
df_sorted_psex['logP'] = -1 * np.log10(df_sorted_psex['stage-sex_p'])
df_sorted_psex['uniform_points'] = range(0, len(df_sorted_psex))
df_sorted_psex['uniform_pval'] = (df_sorted_psex['uniform_points'] + 1) / len(df_sorted_psex)
df_sorted_psex['uniform_logP'] = -1 * np.log10(df_sorted_psex['uniform_pval'])

df_sorted_psex
```

```python
#side-by-side QQ plots

fig, axs = plt.subplots(1,2)

#same as before:
axs[0].scatter(df_sorted["uniform_logP"], df_sorted["logP"])
axs[0].plot([8,0], [8,0], color = "black")
axs[0].set_xlim([0, 8])
axs[0].set_ylim([0, 10])
axs[0].set_xlabel("Expected -log10(p-value)")
axs[0].set_ylabel("Observed -log10(p-value)")
axs[0].set_title("p vals dev stage transcripts")

#sex-adjusted:
axs[1].scatter(df_sorted_psex["uniform_logP"], df_sorted_psex["logP"])
axs[1].plot([8,0], [8,0], color = "black")
axs[1].set_xlim([0, 8])
axs[1].set_ylim([0, 10])
axs[1].set_xlabel("Expected -log10(p-value)")
axs[1].set_ylabel("Observed -log10(p-value)")
axs[1].set_title("p vals dev stage, adj for sex")

plt.show()
# pretty similar, which was expected (I think?)
```

```python
#again do correction for false discovery rate

#using a 10% false discovery rate
fdr = 0.1/5556
df_fdr_sex = df_sorted_psex[df_sorted_psex['stage_p']<fdr]
df_fdr_sex
```

```python
#now find the overlapping genes after correction

dif_exp_genes = list(df_fdr.index.values)
dif_exp_genes2 = list(df_fdr_sex.index.values)

# hooray for set theory
genes = list(set(dif_exp_genes) & set(dif_exp_genes2))
genes
```

```python
#make a new dataframe for a volcano plot

fig,ax=plt.subplots(figsize=(10,10))

x = df_sorted['bvals']
y = df_sorted['logP']

#identify the 'significant' dots

colors = [] #it took me way too long to figure this out

for dot in df_sorted['stage-sex_p']:
    if dot < fdr:
        colors.append('g')
    else:
        colors.append('b') 
        
ax.scatter(x, y)

ax.set_xlabel('Beta')
ax.set_ylabel('-log10(p-value)')
ax.set_title("Differential Expression by Developmental Stage, Sex Adjusted")

plt.show()
```

# Question 1: Write new file

We want to build a mapping from FlyBase gene IDs to UniProt IDs.

Parse the flybase data file (.txt), looking only at the data from the fields containing the string "DROME".

Parse this file, and output a new file with two tab separated columns, the first containg the FlyBase ID and the second the Uniprot ID (AC).

### Note:

While the file is open, the code below is also going to build the dictionary necessary for question 2

```python
gnames = {}

# open input and output files
with open("fly.txt", 'r') as fs:
    with open('fly_uid.txt', 'w') as output:
        # split according to column
        for line in fs:
            if line.count("DROME")==0:
                continue
            else:
                fields = line.split() #create a list and get rid of new line marks
                if line.count("FBg")==0:
                    continue
                else:
                    gnames[fields[-1]] = fields[-2]
                    output.write(fields[-1] + "\t" + fields[-2] + "\n") #writes to a new file output
```

# Question 2: Identifier Mapping Using a Dictionary

Write a function that uses three arguments: 
1. one which will use a referencing mapping file (the 'fly.txt' file above). 
1. another which uses the c_tab file, and replace the *gene_id* fields with the *UniProt ID* from the first file
1. a third which - depending on its value - will do something different for a missing gene_id in the c_tab file:
 - one value shoud cause the function to print nothing (ignore the line) if gene_id is missing
 - another which will print a default value to replace the gene_id field if it is missing
 
NOTE: We only want the first 100 lines of output displayed in your notebook or in an output file


```python
# create new function for mapping gene ID
def mapping_func(file1, file2, remove_no_gID=True): 
    counter = 0 # starts count at 0 (to count 100 lines of the ctab file)    
    file_to_map = open(file1,'r')
    ref_file = open(file2, 'r')

    for line in file_to_map: #reads line by line in the ctab file
        if 't_id' in line: # print title row, unchanged
            print(line)
        else:
            line=line.split('\t') # separate each line into new lines by tab
            if counter>=100: #ends after 100 lines of the file
                break
            elif line[8] in gnames: # what to do if gene_id is present
                line[8]=gnames[line[8]] #replaces gene_id fields with UniProt_ID
                counter+=1
            else: # what to do if gene_id is missing
                if not remove_no_gID:
                    line[8]="No_gene_id"
                    counter+=1
                else: # if the 3rd argument of the function tells us to skip
                    continue
        printline="\t".join(line)
        print(printline)
        
    file_to_map.close()
    ref_file.close()
```
